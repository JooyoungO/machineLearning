**인공 신경망**

7장에서 사용한 밀집층에는 뉴런마다 입력 개수만큼의 가중치가 있어, 모든 입력에 가중치를 곱함

인공 신경망은 처음에 가중치들고 절편을 랜덤하게 초기화한 다음 에포크를 반복하면서 경사 하강법 알고리즘을 사용하여 손실이 낮아지도록 최적의 가중치와 절편을 찾아감

이건이 바로 모델 훈렴임

**합성곱 신경망**

합성곱은 밀집층의 계산과 조금 다름

입력 데이터 전체에 가중치를 적용하는 것이 아니라 일부에 가중치를 곱함

예를들어, 입력값이 10개이고 뉴런이 3개의 가중치를 가지고 있다면 1, 2, 3번째 입력에 대한 뉴런을 수행하고 다음에는 2, 3, 4번째 입력에 대해서 반복

반복 결과 8개의 출력이 나오게 되며, 여기서 사용하는 하나의 뉴런의 가중치와 절편은 동일

합성곱 신경망에서는 완전 연결 신경망(7장에서 만든 신경망)과 달리 뉴런을 필터(filter) 또는 커널(kernel)이라고 부름

이 책에서는 필터는 뉴런 개수를 표현할 때 사용되며, 커널을 입력에 곱하는 가중치를 뜻함

합성곱은 1차원이 아니라 2차원 입력에서도 적용할 수 있는 장점이 있음

입력이 2차원이면 필터도 2차원이어야 함

특성 뱁: 합성곱 계산을 통해 얻은 출력으로 활성호 함수가 적용된 값

밀집층에서 여러개의 뉴런을 사용하듯 합성곱 층에서도 여러 개의 필터를 사용할 수 있음

여러개의 필터를 사용하면 만ㄷ르어진 특성 맵은 순서대로 차곡차곡 쌓임

(2, 2) 크기의 특성맵을 쌓으면 (2, 2, 3) 의 3차원 배열이 됨

밀집층에 있는 뉴런의 가중치가 모두 다르듯이 합성곱 층에 있는 필터의 가중치도 모두 다름

**밀집층 vs 합성곱**

계산은 밀집층과 동일하게 단순히 입력과 가중치를 곱하고 절편을 더한 것

차이는
  1. 2차원 형태를 유지

  2. 입력보다 훨씬 작은 크기의 커널을 사용

이렇게 2차원 구조를 그대로 사용하기 때문에 합성곱 신경망이 이미지 처리 분야에서 뛰어난 성능 발휘

**케라스 합성곱 층**

위와 같이 합성곱은 Conv2D 클래스로 제공됨

Conv2D의 첫번째 매개변수는 필터의 개수

참고로 특성맵은 활성화 함수가 적용된 값

**합성곱 신경망의 정의**

1개 이상의 합성곱 층을 쓴 인공 신경망을 의미함

클래스에 대한 확률을 계산하려면 마지막 층에 클래스 개수만큼 뉴런을 가진 밀집층을 두는 것이 일반적임

**합성곱 신경망 구조에서 자주 사용되는 기능을 알아보자**

**패딩**

커널 크기를 고정하면서도 출력의 크기를 입력과 동일하게 만들기 위해 입력 배열 주위를 가상의 원소로 채우는 것

실제 입력값은 아니므로 0으로 채워지며 계산에 영향을 주지 않음

**세임 패딩(same padding)**

입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩하ㅡㄴ 것

합성곱 신경망에서는 세임 팯ㅇ이 많이 사용됨

**밸리드 패딩(valid padding)**

패딩없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 것

특성맵이 크기가 줄어듦

**왜 합성곱에서는 패딩을 즐겨 사용 ✨**

패딩이 없다는 모서리값들은 커널 도장에 딱 한번만 쓰이지만 다른 원소들으 2번 이상 커널과 계산됨

즉, 모서리에 있는 중요한 정보가 특성 맵으로 잘 전달되지 않을 가능성이 높고 가운데 있는 정보는 두드러지게 표현될 것임

패딩을 하지 않을 경우, 중앙과 모서리 픽셀이 합성곱에 참여하는 비율이 4 : 1 로 크게 차이가 남

반면에 패딩을 한다면 9 : 4 로 차이가 감소함

**스트라이드(stride)**

합성곱 연산에서 커널이 이동하는 칸 수

기본값은 1로 좌우, 위아래로 한칸 씩 이동함

이동하는 크기를 (1, 1)과 같이 튜플을 사용해 각각 지정 가능

합성곱 연산횟수와 특성 맵의 크기를 줄이는 역할을 하나, 자주 사용되지는 않음

**풀링(pooling)**

합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할을 함

그러나 특성 맵의 개수를 줄이지 않음

합성곱처럼 입력 위를 지나가면서 도장을 찍으나 가중치가 없으며 겹치지 않음

**최대 풀링(max pooling)**

도장을 직는 영역에서 가장 큰 값을 선택

**평균 풀링(average pooling)**

도장을 찍는 영역의 평균값을 계산함

**풀링을 사용하는 이유**

팝성곱에서 스트라이드를 크게 하여 특성 맵을 줄이는 것보다 풀링 층에서 쿠기를 줄이는 것이 경험적으로 더 나은 성능을 내기 때문

따라서 합성곱 신경망은 합성곱 층에서 특성  맵을 생성하고 풀링에서 크기를 줄이는 구조가 쌍을 이룸

합성곱 층과 뚜렷이 구분되기 때문에 풀링 층이라 부름

예제 (4, 4) 크기의 특생 맵이 (2, 2) 최대 풀링 적용

**합성곱 신경망의 전체 구조**

**컬러 이미지를 사용한 합성곱**

컬러 이미지는 RGB 채널로 구성되기 때문에 3차원 배열로 표시

하나의 컬러 이미지는 너비와 높이 차원 외에 깊이(또는 채널) 차원이 있음

예를 들어 흑백 이미지가 (4, 4) 라면 컬러 이미지는 (4, 4, 3)

깊이가 있는 입력에서 합성곱을 수행하기 위해서는 커널로 깊이가 필요함

즉, 필터의 커널 크기가 (3, 3) 이 아니라 (3, 3, 3)

커널 배열의 깊이는 항상 입력의 깊이와 같음

따라서 입력과 필터의 차원이 몇 개인지 상관없이 항상 출력은 하나의 값임

케라스 합성곱 층은 입력이 3차원일 것을 기대하므로 2차원 이미지는 깊이 차원이 1이 3차원 배열로 변환

**이미지 분석을 위한 합성곱 신경망 모델 이용**

합성곱 신경망에서 필터는 이미지에 있는 어떤 특징을 찾는다고 생각할 수 있음

처음에는 간단한 기본적인 특성(직선 곡선 등)을 찾고 층이 깊어질수록 다양하고 구체적인 특징을 감지할 수 있도록 필터의 개수를 늘림

또한 어떤 특징이 이미지의 어느 위치에 놓이더라도 쉽게 감지할수 있도록 너비와 높이 차원을 점점 줄여나감
