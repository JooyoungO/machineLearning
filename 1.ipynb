주어진 입력 데이터 포인트에 대해, 훈련 데이터셋에서 가장 가까운 거리에 있는 k개의 데이터 포인트를 찾습니다.


이 k값은 n_neighbors로 설정됩니다

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```

```python
for n in range(1, 50):
  kn.n_neighbors = n
  score = kn.score(kn._fit_X, kn._y)
  if score < 1:
    print(n, score)
    break
```

```python
kn.n_neighbors = n - 1
print(n-1, kn.score(kn._fit_X, kn._y))
```
